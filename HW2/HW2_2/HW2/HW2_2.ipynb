```python
import jieba, jieba.posseg
import re, time, json
```

```python
data = {}
num = 1
with open('novel.txt', encoding='utf-8') as file:
    for line in file:
        line_list = re.sub(r'[^\u4e00-\u9fa50-9。]', '', line).split('。')
        for l in line_list:
            words = jieba.posseg.lcut(l)
            have_n = False
            for term in words:
                if term.flag.startswith('n'):
                    have_n = True
                    if num not in data:
                        data[num] = []
                    data[num].append(term.word)

            if have_n:
                if num % 50 == 0:
                    print('目前進度:', num, '行')
                num += 1
    
    json.dump(data, open('data_new.json', 'w'))
    print('Done.')
```

```python
characters = []
with open('relation.txt', 'r', encoding='utf-8') as rf:
    for line in rf:
        people = line.replace('\n','')
        if people not in characters:
            characters.append(people)
# print(characters)
for c in characters:
    print(c)
print(len(characters))
```

```python
with open('data_new.json', 'r') as file:
    jsonData = json.load(file)
```

```python
charactersDic = {}
for sentc_num in jsonData:
    for word in jsonData[sentc_num]:
        if word in characters:
            if word not in charactersDic:
                charactersDic[word] = []
                charactersDic[word].append(sentc_num)
            else:
                if sentc_num not in charactersDic[word]:
                    charactersDic[word].append(sentc_num)
            
json.dump(charactersDic, open('invertedIndexMap_new.json', 'w'))
```

```python
with open('invertedIndexMap_new.json', 'r', encoding='utf-8') as file:
    invertedIndexMap = json.load(file)
```

```python
outputDic = {}
for c1 in range(len(characters) - 1):
    for c2 in range(c1+1, len(characters)):
        key1 = characters[c1]+" "+characters[c2]
        key2 = characters[c2]+" "+characters[c1]
        if characters[c1] in invertedIndexMap and characters[c2] in invertedIndexMap:
            matchList = list(set(invertedIndexMap[characters[c1]]).intersection(set(invertedIndexMap[characters[c2]])))
            if len(matchList):
                outputDic[key1] = matchList
                outputDic[key2] = matchList
                
json.dump(outputDic, open('relation_new.json', 'w'))
```

```python
tempList = []
for i in outputDic:
    tempList.append((i, len(outputDic[i])))
sortedList = sorted(tempList, key=lambda x: x[1], reverse=True)

rankingDic = {}
for i in sortedList:
    rankingDic[i[0]] = i[1]
    print(i)
```

## Average Degree

```python
count=0
for k in relationDict:
    count+=len(relationDict[k])
average_degree = count/len(relationDict)
```

## Average clustering coefficient

```python
# Average Clustering Coefficient(ACC)
ACC=0

for k in relationDict:
    node_count=len(relationDict[k])
    total_TCenter=node_count*(node_count-1)/2
    
#     center_list = data[k]
    total_TConnect=0
    for i in relationDict[k]:
        for i_friend in relationDict[i]:
            if i_friend != k and i_friend in relationDict[k]:
                total_TConnect+=1
    total_TConnect/=2
    if total_TCenter!=0:
        ACC+=(total_TConnect/total_TCenter)
#         print((total_TConnect/total_TCenter))
ACC/=len(relationDict)            
```

```python
print(ACC)
```

## Diameter

```python
import collections

def bfs(graph, root):
    seen, queue = set([root]), collections.deque([root])
#     print('seen:',seen)
#     print('queue:',len(queue),'\n')
    popLen=0
    dist=0
    while queue:
        dist+=1
        popLen=len(queue)
        for i in range(popLen):
            vertex = queue.popleft()
            for node in graph[vertex]:
                if node not in seen:
                    
                    seen.add(node)
                    queue.append(node)        
    
    
    return dist
    
if __name__ == '__main__':
    diameter=0
    for k in relationDict:
        dist=bfs(relationDict, k)
        if diameter<dist:
            diameter=dist
            print(diameter)
#     print(diameter)
```
